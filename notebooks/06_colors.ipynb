# Colors and palettes

A quick visual tour of every color spec plotlet accepts. Run all cells —
palettes display themselves as swatches (hover a swatch for its hex value).

```python
import plotlet as pt
```

## The default cycle

Artists without an explicit color walk the tab10 cycle. `"C0"`, `"C1"`, …
address it directly and wrap past `"C9"` (`"C10"` is `"C0"` again).

```python
pt.colors
```

## Single-color shortcuts

Everywhere a color is accepted (`color=`, `fill=`, palette values, …), these
forms resolve:

| form | example | notes |
| --- | --- | --- |
| cycle shortcut | `"C0"`, `"C11"` | tab10, wraps |
| named | `"red"`, `"olive"` | tab10-flavored, not CSS |
| single letter | `"b" "g" "r" "c" "m" "y" "k" "w"` | tab10 hue of the same family |
| grayscale string | `"0.5"` | `"0"` black … `"1"` white |
| float tuple | `(0.2, 0.4, 0.6)` | `(r, g, b)` or `(r, g, b, a)` in [0, 1] |
| hex / CSS | `"#aec7e8"`, `"rebeccapurple"` | passes through — SVG speaks CSS |

```python
from plotlet.draw import resolve_color, Palette

specs = ["C0", "C11", "red", "m", "0.5", (0.2, 0.4, 0.6), "#aec7e8", "rebeccapurple"]
Palette(resolve_color(s) for s in specs)
```

## Qualitative palettes

`pt.palette(name)` returns the palette as a list of hex strings — no
matplotlib required. `n` truncates or cycles past the end
(`pt.palette("Set2", 3)`), and a `"_r"` suffix reverses.

```python
pt.palette("Set2")
```

```python
from IPython.display import display

for name in pt.list_palettes():
    print(name)
    display(pt.palette(name))
```

## Continuous colormaps — the full matplotlib set

Every matplotlib colormap is vendored (including `_r` reversed variants):
perceptually-uniform, Brewer sequential and diverging, cyclic, and the
classics. `pt.palette(name, n)` samples any of them into a discrete list —
evenly spaced, endpoints included — with no matplotlib dependency.

```python
len(pt.list_colormaps()), pt.list_colormaps()[:10]
```

```python
pt.palette("viridis", 12)
```

```python
groups = {
    "Perceptually uniform": ["viridis", "plasma", "inferno", "magma", "cividis"],
    "Sequential":           ["Blues", "Greens", "Oranges", "Purples", "Greys",
                             "YlGnBu", "YlOrRd", "PuBuGn", "cubehelix"],
    "Diverging":            ["coolwarm", "RdBu_r", "PiYG", "BrBG", "PuOr",
                             "Spectral", "seismic"],
    "Cyclic":               ["twilight", "twilight_shifted", "hsv"],
    "Classics & misc":      ["turbo", "jet", "rainbow", "terrain", "ocean"],
}
for group, names in groups.items():
    print(f"\n=== {group} ===")
    for name in names:
        print(name)
        display(pt.palette(name, 24))
```

## Palettes on charts

Artists' `palette=` accepts a palette name, a color list, or a
`{category: color}` dict. It colors the levels of a grouping aes
(`fill=<col>` / `color=<col>`) in order of appearance.

```python
df = {"quarter": ["Q1", "Q2", "Q3", "Q4"] * 3,
      "value":   [12, 15, 11, 17,  9, 13, 14, 16,  7, 8, 10, 12],
      "series":  ["north"] * 4 + ["south"] * 4 + ["west"] * 4}

c = pt.chart(data_width=320, data_height=200,
             title='palette="Set2"', ylabel="$M", legend=True)
c.bar(data=df, x="quarter", y="value", fill="series", position="dodge",
      palette="Set2")
c.legend()
```

```python
c = pt.chart(data_width=320, data_height=200,
             title="dict palette — explicit category mapping", ylabel="$M",
             legend=True)
c.bar(data=df, x="quarter", y="value", fill="series", position="dodge",
      palette={"north": "C0", "south": "olive", "west": "#e78ac3"})
c.legend()
```

## Continuous color on charts

Numeric color columns map through `cmap=` (any name from
`pt.list_colormaps()`); `vmin` / `vmax` / `norm` / `center` control the
value → color mapping. See `docs/API.md` for the full contract.

```python
import math

xs = [i / 6 for i in range(60)]
ys = [math.sin(x) for x in xs]

sc = pt.chart(data_width=320, data_height=200,
              title='cmap="viridis" — color follows y', xlabel="x", ylabel="sin(x)")
sc.scatter(data={"x": xs, "y": ys, "value": ys},
           x="x", y="y", color="value", cmap="viridis", size=4)
sc
```
