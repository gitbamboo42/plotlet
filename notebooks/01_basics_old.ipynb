# plotlet in a Jupyter notebook (chained `pt.figure()` API)

This notebook exercises the chained `pt.figure()` API. The recommended API for new work is `pt.chart()` — see `01_basics.ipynb`. plotlet records artist calls and emits one self-contained SVG when the figure auto-renders.

```python
import plotlet as pt
import math

xs = [i * 0.1 for i in range(64)]

(pt.figure()
   .plot(xs, [math.sin(x) for x in xs], label="sin(x)")
   .plot(xs, [math.cos(x) for x in xs], label="cos(x)", linestyle="--")
   .xlabel("x").ylabel("y").title("Hello from plotlet")
   .grid(True).legend())
```

Notice we didn't call `.show()` — when the last expression in a cell is a `Figure`, Jupyter's `_repr_html_` protocol auto-renders it. Same as plotly / pandas DataFrames.

`.show()` works too if you want to render mid-cell:

```python
fig = pt.figure(width=500, height=300)
fig.bar(["A", "B", "C", "D"], [4, 7, 2, 9], color="C2")
fig.title("explicit show()").show()
```

## With numpy / pandas
Any object with `.tolist()` is auto-converted, so numpy arrays and pandas Series Just Work.

```python
import random
random.seed(0)

xs = [i * 0.1 for i in range(120)]
ys = [math.sin(x) + random.gauss(0, 0.15) for x in xs]

(pt.figure()
   .scatter(xs, ys, s=25, alpha=0.6, label="noisy sin")
   .xlabel("x").ylabel("y")
   .legend().grid(True))
```

## Histogram

```python
data = [random.gauss(0, 1) for _ in range(2000)]
pt.figure().hist(data, bins=30, color="C3").title("normal(0, 1)")
```
